```python
from bs4 import BeautifulSoup

# Replace 'yourfile.html' with the path to your HTML file
with open('/home/ach/Downloads/NANOGPT_DEMO/query_docs/10.1021_acsnano.0c09563.txt', 'r', encoding='utf-8') as file:
    html_content = file.read()

# Parse the HTML content
soup = BeautifulSoup(html_content, 'lxml')

# Get all text from the HTML content
text = soup.get_text()

# Remove empty lines
cleaned_text = "\n".join([line.strip() for line in text.splitlines() if line.strip()])

# Print the cleaned text
print(cleaned_text)

# Optional: Save the cleaned text to a file
with open('cleaned_text.txt', 'w', encoding='utf-8') as text_file:
    text_file.write(cleaned_text)
```
